# Popularity Based Recommender System

```python
 ratings_with_name=ratings.merge(books,on='ISBN')
```

```python
ratings_with_name
```

```python
num_rating_df= ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns = {'Book-Rating':'num-rating'}, inplace = True)
num_rating_df
```

```python
avg_rating_df=ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
avg_rating_df.rename(columns = {'Book-Rating':'avg-rating'}, inplace = True)
avg_rating_df
```

```python
popular_df=num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df=popular_df[popular_df['num-rating']>=250].sort_values('avg-rating',ascending=False)
```

```python
popular_df=popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num-rating','avg-rating']]
```

# collaborative filering based recommender system

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
padhe_likhe_user= x[x].index
```

```python
filtered_rating=ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_user)]
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_rating=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
pt =final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score =cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
def recommend(book_name):
    #index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items=sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:11]
    
    for i in similar_items:
        print(pt.index[i[0]])
```

```python
recommend('Dark Angel')
```
